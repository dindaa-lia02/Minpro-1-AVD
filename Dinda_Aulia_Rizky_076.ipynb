# **Data Preparation**

**MISSING VALUES**

```python
missing_values = (df.isna().sum() / len(df)) * 100

if missing_values.sum() > 0:
    print("Terdapat nilai yang hilang dalam dataset:")
    print(missing_values[missing_values > 0])
else:
    print("Tidak ada nilai yang hilang dalam dataset.")

missing_values_df = pd.DataFrame({'Kolom': missing_values.index, 'Persentase Missing': missing_values.values})
display(missing_values_df)
```

**DUPLICATED VALUES**

```python
duplicate_rows = df[df.duplicated()]

if not duplicate_rows.empty:
    print(f"Terdapat {len(duplicate_rows)} baris duplikat dalam dataset.")
else:
    print("Tidak ada nilai duplikat dalam dataset.")

display(duplicate_rows)
```

**OUTLIERS**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)

```

**INCOSISTENT VALUE**

```python
#Cek Nilai Tidak Konsisten
# Misalnya, kita periksa jika ada nilai negatif yang tidak masuk akal dalam kolom numerik
for col in df.select_dtypes(include=['number']).columns:
    if (df[col] < 0).any():
        print(f"\nKolom {col} memiliki nilai negatif yang tidak konsisten:")
        print(df[df[col] < 0])

# Jika ada kolom kategori, kita bisa melihat nilai uniknya untuk mendeteksi ketidakkonsistenan
for col in df.select_dtypes(include=['object']).columns:
    print(f"\nNilai unik pada kolom '{col}':")
    print(df[col].unique())
```

Melakukan penanganan terhadap data pada kolom daily revenue yang tidak konsisten dikarenakan terdapat nilai negatif. Solusinya yakni mengganti dengan nilai median

```python
median_value = df["Daily_Revenue"].median()
df.loc[df["Daily_Revenue"] < 0, "Daily_Revenue"] = median_value
```
