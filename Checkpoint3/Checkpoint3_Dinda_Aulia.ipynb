# **Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Menampilkan data coffee shop dan sumber data**

```python
df = pd.read_csv('/content/datacoffe_shop_bersih (1).csv')
```

```python
print("Sumber Data: https://www.kaggle.com/datasets/himelsarder/coffee-shop-daily-revenue-prediction-dataset")
df.head()
```

# **Menentukan Jam Operasional Optimal dengan Melihat Tren Kunjungan Pelanggan.**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x=df['Operating_Hours_Per_Day'], y=df['Number_of_Customers_Per_Day'], marker='o', color='purple', ci=None)
plt.title('Tren Jumlah Pengunjung per Hari')
plt.xlabel('Jam Operasional per Hari')
plt.ylabel('Jumlah Pengunjung')
plt.grid()
plt.show()
```

 **Insight Berdasarkan Grafik**

1.   Grafik menunjukkan bahwa jumlah pengunjung tidak selalu meningkat seiring dengan bertambahnya jam operasional.
2. Beberapa titik menunjukkan lonjakan jumlah pengunjung, misalnya pada jam 6, 9, dan 14 jam operasional. Hal ini bisa mengindikasikan adanya jam sibuk tertentu, seperti jam sarapan, makan siang, atau sore hari.   
3. Terlihat ada titik penurunan jumlah pengunjung setelah mencapai puncak, misalnya setelah jam 6 dan jam 10. Bisa jadi karena waktu-waktu tersebut tidak terlalu menarik bagi pelanggan, atau ada faktor eksternal seperti cuaca dan promosi yang memengaruhi.



 **Tindakan yang dapat dilakukan**



1.   Promosi dan strategi pemasaran bisa difokuskan pada jam-jam dengan potensi peningkatan pelanggan.
2. Penyesuaian jumlah staf bisa dilakukan untuk meningkatkan efisiensi operasional di jam sibuk dan menghemat biaya di jam sepi.
3. Kedai kopi dapat mempertimbangkan untuk menyesuaikan jam operasional berdasarkan tren jumlah pengunjung.



# **Melihat sebaran pendapatan harian**

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Daily_Revenue'], bins=7, kde=True, color='blue')
plt.title('Distribusi Pendapatan Harian')
plt.xlabel('Pendapatan Harian')
plt.ylabel('Frekuensi')
plt.show()
```

**Insight dan Tindakan yang diambil Berdasarkan Grafik**

1. Distribusi Positif Skewed
* Mayoritas pendapatan harian berada dalam kisaran yang lebih rendah (0 - 2000).
* Hanya sedikit hari yang memiliki pendapatan tinggi (di atas 4000).
* Distribusi ini menunjukkan bahwa bisnis lebih sering mengalami pendapatan rendah daripada tinggi.

Tindakan: Perbanyak strategi yang berhasil dari hari-hari dengan pendapatan tinggi untuk diterapkan di hari lainnya. Fokus pada peningkatan pendapatan harian rata-rata dengan strategi pemasaran yang lebih efektif.

2. Pendapatan Terbanyak dalam Rentang 1000 - 2000
* Sebagian besar hari memiliki pendapatan dalam kisaran ini.
* Rentang ini bisa menjadi referensi untuk menetapkan target minimal pendapatan harian.

Tindakan: Gunakan target ini sebagai dasar perencanaan keuangan dan operasional. Optimalkan jam operasional dan tenaga kerja untuk memastikan pendapatan minimal ini tercapai setiap hari.

3. Hari dengan Pendapatan Sangat Tinggi Jarang Terjadi
* Hanya sedikit hari yang memiliki pendapatan lebih dari 4000.
* Lonjakan ini mungkin disebabkan oleh faktor tertentu seperti diskon, event, atau hari libur.

Tindakan: Identifikasi faktor yang menyebabkan lonjakan pendapatan (promo khusus, event, atau momen tertentu).

# **Korelasi Antar Variabel Numerik**

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['Number_of_Customers_Per_Day', 'Marketing_Spend_Per_Day', 'Daily_Revenue']].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

**Insight dan Tindakan yang diambil Berdasarkan Grafik**

1. Korelasi sebesar 0.74 menunjukkan bahwa jumlah pelanggan memiliki hubungan yang cukup kuat dengan pendapatan harian. Ini berarti semakin banyak pelanggan yang datang, semakin tinggi pendapatan yang dihasilkan.

Tindakan:  Fokus pada layanan pelanggan untuk meningkatkan retensi pelanggan dan mendapatkan lebih banyak pelanggan setia.

2. Korelasi 0.25 menunjukkan bahwa ada sedikit hubungan positif antara pengeluaran pemasaran dan pendapatan, tetapi tidak terlalu signifikan. Ini bisa berarti bahwa strategi pemasaran saat ini belum terlalu efektif dalam meningkatkan pendapatan.

Tindakan: mencoba berbagai metode pemasaran seperti iklan digital (Google Ads, Facebook Ads) atau program referral untuk melihat strategi mana yang lebih efektif.

3. Korelasi 0.03 menunjukkan bahwa peningkatan anggaran pemasaran tidak berdampak signifikan pada jumlah pengunjung. Ini bisa berarti bahwa strategi pemasaran saat ini tidak menjangkau calon pelanggan yang tepat atau ada faktor lain yang lebih berpengaruh terhadap jumlah pengunjung.

Tindakan: Revisi strategi pemasaran, mungkin iklan yang dilakukan tidak menarik atau tidak sesuai dengan target pasar. Alihkan anggaran pemasaran ke strategi yang lebih efektif, seperti pemasaran berbasis komunitas, promosi media sosial, atau influencer marketing. Gunakan strategi word-of-mouth dengan memberikan insentif bagi pelanggan yang membawa teman atau keluarga mereka.
